```python
import pandas as pd
from pymongo import MongoClient
from config import DB_USERNAME, DB_PASSWORD
```

```python
def _connect_mongo():
    """ A util for making a connection to mongo """
    
    mongo_uri = f'mongodb+srv://{DB_USERNAME}:{DB_PASSWORD}@cluster0.xckv1my.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0'
    conn = MongoClient(mongo_uri)

    return conn['Cluster0']


def read_mongo(collection):
    """ Read from Mongo and Store into DataFrame """

    # Connect to MongoDB
    db = _connect_mongo()

    # Make a query to the specific DB and Collection
    cursor = db[collection].find({ 'responses': {"$exists": True} })

    # Expand the cursor and construct the DataFrame
    df =  pd.DataFrame(iter(cursor))

    return df
```

```python
df = read_mongo(collection='trials')
df = df.explode('responses')
df[['subject', 'age', 'sex', 'blocks', 'trial_num', 'response']] = pd.json_normalize(df.responses)

df
```
